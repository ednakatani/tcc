## Base de Dados

Nessa seção é feita a importação da base de dados que será utilizada.

```python
#open csv
df = pd.read_csv("real_data.csv", delimiter=';')

df1 = pd.read_csv("data.csv", delimiter=',', header=None)
```

```python
# iterate over lines of df
for index, row in df.iterrows():
    #print(row['Paciente'])
    
    for index1, row1 in df1.iterrows():
        if row1[0] == row['Paciente']:
            df1.at[index, 'Alterado'] = row['Alterado']
```

## Divisão dos Dados

Nessa seção os dados são divididos em dados de treino e dados de teste.

```python
# split x and y
x = df1.drop('Alterado', axis=1)
y = df1['Alterado']
```

```python
#remove last row of data
train_x = x[:-7]
train_y = y[:-7]

#get the first row of data
test_x = x[-7:]
test_y = y[-7:]
```

## Criação do Modelo

```python
model = KNeighborsRegressor(n_neighbors=3)  
```

```python
model.fit(train_x, train_y)
```

## Predição

```python
prediction = model.predict(test_x)
print(prediction)
```
